```python
import requests
from tqdm import tqdm
import pandas as pd
```

# Download files from Zenodo

NB: You can download directly from Zenodo or using Zenodo API.

* Zenodo API: [API doc](https://developers.zenodo.org)

```python
# Get the file content
dataset_file = "scored-submissions-2020-to-2023.ndjson"
response = requests.get(f"https://zenodo.org/records/10516043/files/{dataset_file}?download=1", stream=True)

# Save file to local
with open(dataset_file, mode="wb") as file:
    for chunk in tqdm(response.iter_content(chunk_size=10*1024)):
        file.write(chunk)
```
